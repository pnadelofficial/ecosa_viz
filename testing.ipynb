```python
df['Policy Domain'] = df.apply(lambda x: x['Policy Domain'].split(', '), axis=1)# .explode()# .value_counts()
df = df.explode('Policy Domain')
```

```python
import re
df['Sector'] = df.apply(lambda x: re.sub('\(.*\)', '', x['Sector']).split(', '), axis=1)
df = df.explode('Sector')
```

```python
df['Parties'] = df['Parties'].str.split(', ')
df = df.explode('Parties')
```

```python
df
```

```python
policy_domain = []
form_of_cooperation = []
sector = []
country = []
year = []
```

```python
country = [
    "Belarus (Europe)"
]
```

```python
if policy_domain == []:
    policy_domain = list(df['Policy Domain'].unique())
if form_of_cooperation == []:
    form_of_cooperation = list(df['Form of Cooperation'].unique())
if sector == []:
    sector = list(df['Sector'].unique())
if country == []:
    country = list(df['Parties'].unique())
if year == []:
    year = list(df['Date'].dt.year.unique())
```

```python
filtered_data = df[
    (df['Policy Domain'].isin(policy_domain)) &
    (df['Form of Cooperation'].isin(form_of_cooperation)) &
    (df['Sector'].isin(sector)) &
    (df['Parties'].isin(country)) &
    (df['Date'].dt.year.isin(year))
]
```
